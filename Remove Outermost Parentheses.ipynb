```python
s = "(()())(())"
balance=0
lst=[]
new_lst=[]
for i in s:
    if (i=="("):
        balance+=1
        lst.append(i)
    elif(i==")"):
        balance-=1
        lst.append(i)
    if(balance==0):
        joined="".join(lst)
        new_lst.append(joined)
        lst=[]
newest=[]
for k in new_lst:
    length=len(k)
    sliced=k[1:length-1]
    newest.append(sliced)
joined2="".join(newest)
print(joined2)
```
