```python
students = [1,1,1,0,0,1]
sandwiches = [1,0,0,0,1,1]
count0=students.count(0)
count1=students.count(1)
print(count0,count1)
for i in range(len(sandwiches)):
    if(sandwiches[i]==1) and (count1>=1):
        count1-=1
    elif(sandwiches[i]==0) and (count0>=1):
        count0-=1
    else:
        break
final_count=count1+count0

print(final_count)
```
