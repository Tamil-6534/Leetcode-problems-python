```python
s= 'successes'
vowels= ["a","e","i","o","u"]
consonants=['b','c','d','f','g','h','j','k','l','m','n','p','q','r','s','t','v','w','x','y','z']
vowel_dict={"a":0,"e":0, "i":0,"o":0,"u":0}
consonant_dict={'b':0,'c':0,'d':0,'f':0,'g':0,'h':0,'j':0,'k':0,'l':0,'m':0,'n':0,'p':0,'q':0,'r':0,'s':0,'t':0,'v':0,'w':0,'x':0,'y':0,'z':0}
vowel_length=len(vowels)
consonant_length=len(consonants)
for vowel_and_consonant in s:
    for vowel2 in range(vowel_length):
        if(vowel_and_consonant==vowels[vowel2]):
            vowel_dict[vowel_and_consonant] +=1

    for consonant2 in range(consonant_length):
        if(vowel_and_consonant==consonants[consonant2]):
            consonant_dict[vowel_and_consonant] +=1
max_vowel=(max(vowel_dict.values()))
max_consonant=max(consonant_dict.values())
sum=max_vowel+max_consonant
print(sum)
```
