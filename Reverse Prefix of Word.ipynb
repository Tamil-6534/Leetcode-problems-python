```python
word = "abcdefd"
ch = "d"
reversed=[]
for i in word:
    if(i==ch):
        index=word.index(ch)
        for j in word[index::-1]:
            reversed.append(j)
        for k in word[index+1:]:
            reversed.append(k)
        joined=''.join(reversed)
        print(joined)    
        break
if(ch not in word):
    print(word)
```
