```python
s = "((((()(()()()*()(((((*)()*(**(())))))(())()())(((())())())))))))(((((())*)))()))(()((*()*(*)))(*)()"

min_open = 0  # minimum possible '('
max_open = 0  # maximum possible '('

for ch in s:
    if ch == '(':
        min_open += 1
        max_open += 1
    elif ch == ')':
        min_open -= 1
        max_open -= 1
    else:  # '*'
        min_open -= 1      # treat as ')'
        max_open += 1      # treat as '('

    # If too many closing brackets
    if max_open < 0:
        print(False)
        break

    # min_open should never be negative
    if min_open < 0:
        min_open = 0
else:
    print(min_open == 0)
print(min_open,max_open)
```
