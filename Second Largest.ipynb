```python
arr = [12, 35, 1, 10, 34, 1]
arr.sort()
arr.reverse()
big=arr[0]
for i in arr:
    if(big>i):
        print(i)
        break
```

```python
#Second Largest Digit in a String
s = "dfa12321afd"
lst=[int(i) for i in s if i.isdigit()]
second=[]
lst.sort()
lst.reverse()
if(len(lst)!=0):
    big=lst[0]
    for i in lst:
        if(big>i):
            print(i)
            second.append(i)
            break
if(len(second)==0):
    print(-1)


    
```
