```python
allowed = "ab"

words = ["ab","bd","aaab","baa","badab"]
splitted=[allow for allow in allowed]
finded=[]

alphabets=["a",'b','c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z' ]
for i in splitted:
    if(i in alphabets):
        alphabets.remove(i)
for k in alphabets:
    for j in words:
        if(k in j):
            finded.append(j)
for l in finded:
    if(l in words):
        words.remove(l)
length=len(words)
print(length)
```
