```python
class MyStack:
    def __init__(self):
        # Initialize an empty list to represent the stack
        self.stack = []

    def push(self, x: int) -> None:
        """Push element x onto stack."""
        self.stack.append(x)
        # Push usually returns nothing (None)

    def pop(self) -> int:
        """Removes the element on top of the stack and returns that element."""
        if not self.empty():
            return self.stack.pop()
        return None

    def top(self) -> int:
        """Get the top element."""
        if not self.empty():
            return self.stack[-1]
        return None

    def empty(self) -> bool:
        """Returns whether the stack is empty."""
        return len(self.stack) == 0

    def __str__(self):
        """Allows us to see the stack contents when we call print(obj)."""
        return f"Stack: {self.stack}"

# --- EXECUTION LOGIC ---

obj = MyStack()
inputs = ["push", "push", "top", "pop", "empty"]
# Values to be used for the 'push' commands
values = [10, 20, None, None, None]

print("Starting Stack Operations...\n")

for i in range(len(inputs)):
    action = inputs[i]
    val = values[i]
    
    # getattr(obj, action) converts the string "push" into the method obj.push
    method = getattr(obj, action)
    
    if action == "push":
        method(val)
        print(f"Action: {action}({val}) -> State: {obj}")
    else:
        result = method()
        print(f"Action: {action}() -> Result: {result} | State: {obj}")

print("\nFinal Stack State:", obj)
```
