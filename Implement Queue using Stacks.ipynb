```python
class MyQueue:

    def __init__(self):
        self.queue=[]
        

    def push(self, x):
        self.queue.append(x)
        
    def pop(self):
        return self.queue.pop(0)
        

    def peek(self):
        return self.queue[0]
        

    def empty(self):
        return len(self.queue)==0
    
inputs=[ "push", "push", "peek", "pop", "empty"]
values=[[1], [2], [], [], []]
obj=MyQueue()
for i in range(len(inputs)):
    input=inputs[i]
    val=values[i]
    method=getattr(obj,input)

    if(input=='push'):
        print(method(val))
    else:
        print(method())
```
