```python
s = "abBAcC"
stack=[]
length=len(s)
for i in range(length):
    stack.append(s[i])
    if(len(stack) >=2):
        if(s[i]==s[i].lower()):
            if(s[i].upper()==stack[-2] ):
                stack.pop(),stack.pop()
        elif(s[i]==s[i].upper()):
            if(s[i].lower()==stack[-2]):
                stack.pop(),stack.pop()
joined=''.join(stack)
joined

    
```
