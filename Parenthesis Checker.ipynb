```python
s = '{[('
stack =[]
for i in range(len(s)):
    stack.append(s[i])
    if len(stack) >= 2 and stack[-1] == ']' and stack[-2] == '[':
        stack.pop(),stack.pop()
    if len(stack) >= 2 and stack[-1] == '}' and stack[-2] == '{':
        stack.pop(),stack.pop()
    if len(stack) >= 2 and stack[-1] == ')' and stack[-2] == '(':
        stack.pop(),stack.pop()
if(len(stack)==0):
    print('true')
else:
    print('false')
```
