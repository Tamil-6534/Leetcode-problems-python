```python
s = 'a'
first=[]
length=len(s)
countting=s.count(s[0])
if(length==countting) and (len(s)>1):
    print(-1)
else:
    for i in range(len(s)):
        countt=s.count(s[i])
        if(countt==1):
            first.append(i)
            print(first[0])
            break
if(len(first)==0) and (len(s)>1):
    print(-1)
elif(len(s)==1):
    print(0)
```
