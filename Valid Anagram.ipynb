```python
S = list("a")
T = list("ab")
S.sort(),T.sort()
dict1={}
dict2={}
count=0
if(len(S) == len(T)):
    for i in range(len(T)):
        dict1[S[i]]= dict1.get(S[i],0)+1
        dict2[T[i]]=dict2.get(T[i],0)+1
        if(S[i]==T[i]):
            count+=1
if(count==len(S) and len(T)):
    print(True)
else:
    print(False)
```
