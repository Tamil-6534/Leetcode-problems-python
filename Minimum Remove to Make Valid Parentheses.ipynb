```python
s = "())()((("
s = "lee(t(c)o)de)"
s=')(t)h(e(o)'
lst=list(s)
stack_open=[]
stack_close=[]
unwanted=[]
for i in range(len(s)):
    if(lst[i]=='('):
        stack_open.append(i)
    elif(lst[i]==')') :
        if(len(stack_open)!=0):
            stack_close.append(i)
            stack_open.pop()
            stack_close.pop()
        else:
            unwanted.append(i)
for i in stack_open:
    lst.insert(i+1,'0')
    lst.pop(i)
for i in stack_close:
    lst.insert(i+1,'0')
    lst.pop(i)
for i in unwanted:
    lst.insert(i+1,'0')
    lst.pop(i)
joined=''.join(lst)
new=joined.replace('0','')


print(stack_open,stack_close,unwanted)
print(joined)
        
```
